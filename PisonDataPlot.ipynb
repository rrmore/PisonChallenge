```python
# Obtain visualisation of data to get correlation. X axis is timestamp and Y is data. Red dot indicates active label
import time
import psutil
import matplotlib.pyplot as plt
import json
import ijson
import numpy as np
import matplotlib.ticker as ticker
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
```

```python
# Stored part of the data stream in this file and used it to plot graph
F = open("MyFile.txt","r") 
```

```python
actX=[]
actY=[]
restX=[]
restY=[]
i=0
for line in F:
    y = json.loads(line)
    if(y['label']=='REST'):
        restX.append(y['timeStamp'])
        restY.append(y['data'])
    else:
        actX.append(y['timeStamp'])
        actY.append(y['data'])
    ax.scatter(actX,actY,marker='o',color='r')
    ax.scatter(restX,restY,marker='o',color='b')
    ax.yaxis.set_major_locator(ticker.MultipleLocator(3000))
    fig.canvas.draw()
    time.sleep(0.5)
    
```
